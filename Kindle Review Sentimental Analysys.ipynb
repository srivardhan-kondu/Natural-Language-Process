```python
#  data pre processing and cleanning
# postive--1, negitive--0
```

```python
df['rating']=df['rating'].apply(lambda x:0 if x<3 else 1)
```

```python
# 1. LOWER ALL THE CASES
```

```python
df['reviewText']=df['reviewText'].str.lower()
```

```python
df.head()
```

```python
#  Removing special characters
```

```python
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
from bs4 import BeautifulSoup
```

```python
import re
# Step 1: Remove special characters
df['reviewText'] = df['reviewText'].apply(lambda x: re.sub(r'[^a-z A-Z 0-9\s]', '', str(x)))

# Step 2: Remove stopwords
df['reviewText'] = df['reviewText'].apply(lambda x: " ".join([word for word in x.split() if word not in stopwords.words('english')]))

# Step 3: Remove URLs (with extended regex)
df['reviewText'] = df['reviewText'].apply(lambda x: re.sub(r'(http|https|ftp):\/\/[\w_-]+(?:\.[\w_-]+)+(?:[\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '', x))
df['reviewText']=df['reviewText'].apply(lambda x:BeautifulSoup(x,'lxml').get_text())
df['reviewText']=df['reviewText'].apply(lambda x: " ".join(x.split()))

```

```python
df.head()
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
import nltk
nltk.download('wordnet')
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
df['reviewText']=df['reviewText'].apply(lambda x:lemmatize_words(x))
```

```python
df.head()
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df['reviewText'],df['rating'],test_size=0.2,random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow=CountVectorizer()
X_train_bow=bow.fit_transform(X_train).toarray()
X_test_bow=bow.transform(X_test).toarray()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid=TfidfVectorizer()
X_train_tfid=tfid.fit_transform(X_train).toarray()
X_test_tfid=tfid.transform(X_test).toarray()
```

```python
X_train_bow
```

```python
from sklearn.naive_bayes import GaussianNB
model_bow=GaussianNB().fit(X_train_bow,y_train)
model_tfidf=GaussianNB().fit(X_train_tfid,y_train)

```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
y_pred_bow=model_bow.predict(X_test_bow)
y_pred_tfid=model_tfidf.predict(X_test_tfid)
```

```python
print(confusion_matrix(y_test,y_pred_bow))
print(classification_report(y_test,y_pred_bow))
print(accuracy_score(y_test,y_pred_bow))
```
